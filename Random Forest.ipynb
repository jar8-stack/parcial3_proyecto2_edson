# Empezando el algoritmo Random Forest


```python
from matplotlib import pyplot as plt
%matplotlib inline
```

# Variables de procesamiento

```python
tren2
```

```python
tren_test
```

```python
#Preparamos los datos de entrenamiento.
sex = pd.get_dummies(tren2['Sex'],drop_first=True)
#cabin = pd.get_dummies(tren2['Cabin'],drop_first=True)
embarked = pd.get_dummies(tren2['Embarked'],drop_first=True)
```

```python
tren2.drop(['Sex','Embarked','Name','Ticket', 'Cabin'],axis=1,inplace=True)
tren2 = pd.concat([tren2,sex,embarked],axis=1)
tren2 #Datos de entrenamiento
```

```python
#Preparamos los datos para testeo
sex = pd.get_dummies(tren_test['Sex'],drop_first=True)
#cabin = pd.get_dummies(tren_test['Cabin'],drop_first=True)
embarked = pd.get_dummies(tren_test['Embarked'],drop_first=True)
```

```python
tren_test.drop(['Sex','Embarked','Name','Ticket', 'Cabin'],axis=1,inplace=True)
tren_test = pd.concat([tren_test,sex,embarked],axis=1)
tren_test #Datos de testeo
```

```python
X_train= tren2.drop(['Survived','Desconocido'],axis='columns')
Y_train= tren2.Survived
```

```python
X_test = tren_test.drop('Survived',axis='columns')
Y_test = tren_test.Survived
```

```python
X_test
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, Y_train)
```

```python
model.score(X_test, Y_test)
```

```python
Y_predicted = model.predict(X_test)
```

```python
#Matriz de confusion
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_predicted)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn import tree
model.estimators_
```
